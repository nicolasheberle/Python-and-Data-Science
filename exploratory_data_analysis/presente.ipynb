```python
import pandas as pd

# Suponha que temos um DataFrame com as colunas 'Coluna1' e 'Coluna2'
df = pd.DataFrame({
    'Coluna1': ['pao', 'trigo', 'arroz', 'feijao'],
    'Coluna2': ['trigo', 'arroz', 'feijao', 'banana']
})

# Cria uma nova coluna 'Coluna3' que verifica se o valor na 'Coluna1' está presente na 'Coluna2'
df['Coluna3'] = df['Coluna1'].isin(df['Coluna2'])

# Exibe o DataFrame
print(df)
```
