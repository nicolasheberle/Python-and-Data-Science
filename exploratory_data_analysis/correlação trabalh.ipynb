```python
import pandas as pd

# Carregar as planilhas, verificando o separador e nomes das colunas
planilha1 = pd.read_csv('2_andar_16-07-2024.csv', sep=';') # Substitua ';' pelo separador correto se necessário
planilha2 = pd.read_csv('todos_3_andares.csv', sep=';') # Substitua ';' pelo separador correto se necessário

# Imprimir os nomes das colunas para verificar se 'Service Tag' está presente e com a capitalização correta
print(planilha1.columns)
print(planilha2.columns)

# Coluna de interesse para comparação (ajuste o nome se necessário)
coluna_interesse = 'Service Tag'

# Criar uma lista para armazenar as linhas correspondentes
linhas_correspondentes = []

# Iterar sobre cada valor na coluna de interesse da planilha1
for valor in planilha1[coluna_interesse]:
    # Verificar se o valor está presente na planilha2
    if valor in planilha2[coluna_interesse].values:
        # Extrair a linha correspondente da planilha2
        linha_correspondente = planilha2[planilha2[coluna_interesse] == valor]
        # Adicionar a linha à lista de linhas correspondentes
        linhas_correspondentes.append(linha_correspondente)

# Concatenar as linhas correspondentes em um DataFrame
informacoes_correspondentes = pd.concat(linhas_correspondentes)

# Substituir o conteúdo da coluna 'Local' por 'Sede 14º'
informacoes_correspondentes['Local'] = 'Sede 13º'

# Salvar as informações correspondentes em uma nova planilha
informacoes_correspondentes.to_csv('informacoes_correspondentes_2.csv', index=False)

print('As informações correspondentes foram extraídas e salvas em "informacoes_correspondentes_2.csv".')
```
