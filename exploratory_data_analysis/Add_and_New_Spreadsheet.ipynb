### Home-Office
### Estoque CG
### Caldeira
### Indisponível

## Padronizar os tipos de dados.

```python
# Puxar biblioteca
import pandas as pd

# Puxar o dataset
df = pd.read_excel('monitores_att.xlsx')

# Alterar o tipo dele
df.to_csv('monitores_att.csv', index = False)
```

## Agora com ambos os dados no mesmo formato, facilitará o processo de desenvolvimento.

```python
# Carregar a planilha existente
df_existing = pd.read_csv('prontinho.csv')
df_existing = df_existing.rename(columns={'Local': 'Status'})

print(df_existing.head(5))
```

```python
# Carregar a planilha de onde as linhas serão adicionadas
df_to_add = pd.read_csv('monitores_att.csv')

print(df_to_add.head(5))
```

```python
# Filtrar as linhas que possuem as informações desejadas no 'Local'
df_filtered = df_to_add[df_to_add['Status'].isin(['Home-Office', 'Estoque CG', 'Caldeira', 'Indisponível'])]

# Renomeie a coluna 'nome_antigo' para 'nome_novo'
df_filtered = df_filtered.rename(columns={'Serial': 'Service Tag'})

print(df_filtered.head(5))
```

```python
# Concatenar as linhas filtradas à planilha existente
df_final = pd.concat([df_existing, df_filtered])

# Salvar a planilha final
df_final.to_csv('planilha_final_1.csv', index=False)

# Imprimir uma mensagem de sucesso
print("Linhas adicionadas com sucesso à planilha existente.")
```
